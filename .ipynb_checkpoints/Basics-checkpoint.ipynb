```python
import pandas as pd
board_games = pd.read_csv('board_games.csv')
print(board_games.head())
```

```python
board_games.dropna(axis=0)
```

```python
from sklearn.cluster import KMeans
cluster = KMeans(n_clusters=5)
numeric_columns = board_games.drop(board_games.columns[[0,1,2]], axis=1, inplace=True)
```

```python
cluster.fit(numeric_columns)
```
